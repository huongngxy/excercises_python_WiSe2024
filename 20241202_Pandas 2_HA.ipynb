```python
import pandas as pd 
import numpy as np
#import datetime
```

```python
# Read file 
file = "Data/TSLA.csv"
df = pd.read_csv(file)
df = df[['Date','Close']]
df
```

```python
# filter/ select data from 2011 to 2022. Use function dt.year 
# If correct then the shape should be (3020,3) 
df['Date'] = pd.to_datetime(df['Date'], format = "%Y-%m-%d")
#df['Year'] = df['Date'].dt.year

df1 = df[(df['Date'].dt.year >= 2011) & (df['Date'].dt.year <= 2022) ]
df1
```

```python
# calculate logarithmic return with df['ln_Returns']

df1['ln_Returns'] = np.log(df1['Close'] / df1['Close'].shift(1))  # Formula? 
df1
```

```python
# calculate annual mean of logarithmic returns 
df_avg = df1.groupby(df['Date'].dt.year)['ln_Returns'].mean().reset_index()
df_avg
```

```python
# save the average log-return of each year in a dictionary named "my_dictionary", whereas the years are keys and the average are values  
# Use a for loop
# use function zip 

# my_dictionary = dict(zip(df1['Date'], df1['ln_Returns']))
# print(my_dictionary)


my_dictionary = dict(zip(df_avg['Date'], df_avg['ln_Returns'].round(5)))

my_dictionary
```
